```python
#NDOT Splitter - a quick and inefficient code to handle the NDOT RWIS data
#make sure you are in the same directory as the files you want to split
#Copyright: Eric Becerril-Blas
import csv

dates = []
wind_speed = []
wind_direction = []
gust_speed = []
gust_direction = []
visibility = []
air_temp = []
humidity = []
dew_point = []
rain = []
rain_rate = []


#First let's take care of this wind speed--------------------

#Function which takes care of wind speeds
def WindSpeed(fileName, dates, wind_speed, wind_direction, gust_speed, gust_direction):
    with open(fileName, 'r') as h:
        lines_after_Juan = h.readlines()[1:]#Skip the first line
        for x in lines_after_Juan:
            dates.append(x.split(',')[0])
            wind_speed.append(x.split(',')[1])
            wind_direction.append(x.split(',')[2])
            gust_speed.append(x.split(',')[3])
            gust_direction.append(x.split(',')[4])
    h.close()


#Make sure you are in the same directories as these files
fileName = 'February.csv'
with open(fileName, 'r') as h:
    lines_after_Juan = h.readlines()[0:]
    for x in lines_after_Juan:
        dates.append(x.split(',')[0])
        wind_speed.append(x.split(',')[1])
        wind_direction.append(x.split(',')[2])
        gust_speed.append(x.split(',')[3])
        gust_direction.append(x.split(',')[4])
h.close()
#WindSpeed(fileName, dates, wind_speed,wind_direction, gust_speed, gust_direction) #Commented out due to no January data

fileName = 'March.csv'
WindSpeed(fileName, dates, wind_speed,wind_direction, gust_speed, gust_direction)

fileName= 'April.csv'
WindSpeed(fileName, dates, wind_speed,wind_direction, gust_speed, gust_direction)

fileName = 'May.csv'
WindSpeed(fileName, dates, wind_speed,wind_direction, gust_speed, gust_direction)

fileName = 'June.csv'
WindSpeed(fileName, dates, wind_speed,wind_direction, gust_speed, gust_direction)

fileName = 'July.csv'
WindSpeed(fileName, dates, wind_speed,wind_direction, gust_speed, gust_direction)

fileName = 'August.csv'
WindSpeed(fileName, dates, wind_speed,wind_direction, gust_speed, gust_direction)

fileName = 'September.csv'
WindSpeed(fileName, dates, wind_speed,wind_direction, gust_speed, gust_direction)

fileName = 'October.csv'
WindSpeed(fileName, dates, wind_speed,wind_direction, gust_speed, gust_direction)

fileName = 'November.csv'
WindSpeed(fileName, dates, wind_speed,wind_direction, gust_speed, gust_direction)

fileName = 'December.csv'
WindSpeed(fileName, dates, wind_speed,wind_direction, gust_speed, gust_direction)

#print(first_line) #Test Output
#print (dates) #Test Output
#print(wind_speed) #Test Output

#Create new file which puts the data
f = open('WindSpeed2018.csv', 'w')
for i in range(len(dates)):
    f.write(dates[i])
    f.write(',')
    f.write(wind_speed[i])
    f.write(',')
    f.write(wind_direction[i])
    f.write(',')
    f.write(gust_speed[i])
    f.write(',')
    f.write(gust_direction[i])
    f.write('\n')
f.close()
```
